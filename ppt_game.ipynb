```python
import random

opciones = ["piedra", "tijera", "papel"]

nombre_jugador = str(input("Introduce tu nombre:").title())
```

A continuaci贸n, se definen 2 funciones. 

- **`eleccion()`**: Esta funci贸n permite elegir entre piedra, papel o tijera. 
- **`juego_ppt()`**: Esta funci贸n contiene la l贸gica del juego y permite iniciarlo.

```python
def eleccion(jugador):

    if jugador.lower() in opciones:
        return jugador  
    
    else: 
        return eleccion(input("Elige una entrada v谩lida: piedra, papel o tijera"))
    
def juego_ppt():
    
    victorias_1 = 0
    victorias_2 = 0

    while victorias_1 < 3 or victorias_2 < 3:
    
        jugador = eleccion(input("Elige piedra, papel o tijera"))

        machine = eleccion(random.choice(opciones))
    
        if jugador == machine:
            print(f"★ {nombre_jugador} eligi贸 {jugador}\n★ La m谩quina eligi贸 {machine}\n Hay un empate\n")

        elif jugador == "piedra" and machine == "tijera": 
            victorias_1 += 1
            print(f"★ {nombre_jugador} eligi贸 {jugador}\n★ La m谩quina eligi贸 {machine}\nコ Gana {nombre_jugador}\n")
        
        elif jugador  == "tijera" and machine == "papel": 
            victorias_1 += 1
            print(f"★ {nombre_jugador} eligi贸 {jugador}\n★ La m谩quina eligi贸 {machine}\nコ Gana {nombre_jugador}\n")
        
        elif jugador  == "papel" and machine == "piedra": 
            victorias_1 += 1
            print(f"★ {nombre_jugador} eligi贸 {jugador}\n★ La m谩quina eligi贸 {machine}\nコ Gana {nombre_jugador}\n")
        
        else: 
            victorias_2 += 1 
            print(f"★ {nombre_jugador} eligi贸 {jugador}\n★ La m谩quina eligi贸 {machine}\n Gana la m谩quina\n")
        
        if victorias_2 == 2 or victorias_1 == 2:
            break
    
    if victorias_1 > victorias_2:
        print(f"El mejor de 3 ha sido {nombre_jugador} コ\n")
    else:
        print(f"El mejor de 3 ha sido la m谩quina. {nombre_jugador} ha perdido \n")
```

1. Se inicia la partida llamando a la funcion `juego_ppt()`

3. Se pregunta si el jugador quiere seguir jugando. Si es que s铆, se vuelve a iniciar una partida. Si es que no, se para el juego. En caso de que el jugador introduzca algo que no sea "si/no", se vuelve a preguntar si el jugador quiere seguir jugando.
